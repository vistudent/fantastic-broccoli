```python
# 3. Calculating and plotting PSD
import matplotlib.pyplot as plt
from pycbc import frame
#loading the appropriate format
strain = frame.read_frame('challenge2.gwf', 'H1:CHALLENGE2')
from pycbc.psd import interpolate, inverse_spectrum_truncation
#removing 2 seconds of data from both the start and end, like in the method shown in the lecture
conditioned = strain.crop(2, 2)
#using 4 second samples
psd = conditioned.psd(4)
#interpolating
psd = interpolate(psd, conditioned.delta_f)
#filter
psd = inverse_spectrum_truncation(psd, int(4 * conditioned.sample_rate),
                                  low_frequency_cutoff=20)

#plotting psd
psd.plot()
plt.xscale('log')
plt.yscale('log')
plt.title('Power Spectral Density (PSD)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('PSD')
plt.show()

# 4. Calculating and plotting SNR

#resizing, so template and data have the same length
tlen = max(len(conditioned), len(hp))
conditioned.resize(tlen)
hp.resize(tlen)

#code for calculating and plotting SNR taken from the lecture
snr = matched_filter(hp, conditioned,
                     psd=psd, low_frequency_cutoff=20)

snr = snr.crop(4 + 4, 4)
pylab.figure(figsize=[10, 4])
pylab.plot(snr.sample_times, abs(snr))
pylab.ylabel('Signal-to-noise')
pylab.xlabel('Time (s)')
pylab.show()

peak = abs(snr).numpy().argmax()
snrp = snr[peak]
time = snr.sample_times[peak]

# 5. Matched-filter SNR
print("We found a signal at {}s with SNR {}".format(time,
                                                    abs(snrp)))
```

```python
#Part 1
#loading the data and plotting time-series data like shown in the hints
import numpy as np
import pycbc
from pycbc.types import FrequencySeries, TimeSeries
import matplotlib.pyplot as plt
d = np.load('noise_ts_4096Hz.npy')
time = d[:, 0]
strain = d[:, 1]
delta_t = time[1] - time[0]
data = TimeSeries(strain, delta_t=delta_t)
plt.plot(time, data)
plt.xlabel('time (sec)')
plt.ylabel('strain')
```

```python
#Part 1
#Calculating and plotting PSD, just like before

import numpy as np
import pycbc
from pycbc.types import TimeSeries
from pycbc.psd import welch, interpolate, inverse_spectrum_truncation
from pycbc.filter import matched_filter
import matplotlib.pyplot as plt

psd = data.psd(4)
psd = interpolate(psd, data.delta_f)
psd = inverse_spectrum_truncation(psd, int(4 * data.sample_rate), low_frequency_cutoff=30)
plt.loglog(psd.sample_frequencies, psd, label='PSD')
plt.xlabel('Frequency (Hz)')
plt.ylabel('PSD')
plt.legend()
plt.show()

```

```python
#Part 1
#generating template waveforms with mass ranges between 5 to 10 (in an array)
mass_ranges = np.arange(5, 11, 1)
template_snr = []

for mass in mass_ranges:
    #creating template waveform, lower frequency set to 30Hz
    template, _ = pycbc.waveform.get_td_waveform(approximant="SEOBNRv4_opt",
                                              mass1=mass, mass2=mass,
                                              delta_t=delta_t, f_lower=30)

    tlen = max(len(data), len(template)) #resizing
    data.resize(tlen)
    template.resize(tlen)

    #calculating matched filter
    snr = matched_filter(template, data)
    template_snr.append(max(abs(snr)))
```

```python
#as stated in the assignment, if SNR>8, there is a detection
if max(template_snr) > 8:
    print("Gravitational wave signal detected")
else:
    print("No gravitational wave signal detected")
```

```python
# Part 2
import numpy as np
from pycbc.psd import welch, interpolate, inverse_spectrum_truncation
from pycbc.types import TimeSeries

#estimating psd using welch
psd = welch(data)

#interpolating and appling inverse spectrum truncation, like in previous examples
psd = interpolate(psd, data.delta_f)
psd = inverse_spectrum_truncation(psd, int(4 * data.sample_rate), low_frequency_cutoff=30)

#taken from the code from the lecture
psd[0] = psd[1]
psd[len(psd) - 1] = psd[len(psd) - 2]
confd = data.to_frequencyseries()
data_whitened = (confd / psd**0.5).to_timeseries()
data_whitened = data_whitened.crop(4,4)
plot = data_whitened.plot()
```

```python
#Part 2
#histogram for whitened data
plt.hist(data_whitened, bins=50, density=True, alpha=0.75, color='g')
plt.xlabel('Whitened Data Values')
plt.ylabel('Probability Density')
plt.title('Histogram of Whitened Data')
plt.show()
```

```python
#calculating and printing mean and standard deviation of the data
mean_value = np.mean(data_whitened)
std_dev = np.std(data_whitened)
print(mean_value)
print(std_dev)
#values show that data follows gaussian distribution
```
